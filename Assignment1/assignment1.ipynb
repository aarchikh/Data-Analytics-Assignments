 1. Why must dictionary keys be immutable in Python? 

Dictionary keys must be immutable because Python uses hashing to store and
retrieve keys efficiently. If a key changes its value, its hash also changes,
making it impossible for the dictionary to locate it.


```python
d = {(1, 2): "Valid key"}   # Immutable key 
print(d)

# Mutable key 
# d = {[1, 2]: "Invalid key"}   # TypeError
```

2. Write a program that takes a list of elements and creates a dictionary
showing the frequency of each element.


```python
elements = [1, 2, 2, 3]
freq = {}

for item in elements:
    freq[item] = freq.get(item, 0) + 1

print(freq)
```

 3. What is the difference between dict.get(key) and dict[key]?

dict.get(key) - returns the value of the specified key and returns None
  or a default value if the key does not exist, making it safer to use.

dict[key] - returns the value of the specified key but raises a KeyError
  if the key does not exist in the dictionary.

```python
## Example where get() is safer

student = {"name": "Aarchi", "age": 19}

print(student.get("marks", 10))  # Using get() 

# Using dict[key] - unsafe if key does not exist
# print(student["marks"])   # KeyError
```

 4. Write a program to merge two dictionaries and resolve common keys
by keeping the larger value.


```python
dict1 = {"a": 10, "b": 20, "c": 30}
dict2 = {"a": 25, "c": 15, "d": 40}

merged_dict = {}

for key in dict1.keys() | dict2.keys():
    merged_dict[key] = max(dict1.get(key, 0), dict2.get(key, 0))

print(merged_dict)
```

 5. Using a dictionary comprehension, create a dictionary from a list of
integers where keys are numbers and values are their cubes, but only
for odd numbers.


```python
numbers = [1,2, 3, 5, 7,8]
cube = {n: n**3 for n in numbers if n % 2 != 0}
print(cube)
```
