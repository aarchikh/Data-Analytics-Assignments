```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

*1. Explain: Data Cleaning & Preprocessing 
• Load the dataset and check for missing values. 
• Fill missing Age values with median or mean. 
• Handle missing Embarked values by filling with the most common port. 
• Drop unnecessary columns like Name, Ticket, Cabin for visualization purposes.*

=> Data cleaning and preprocessing is the process of preparing raw data for analysis.
It involves handling missing values, correcting data types, and removing unnecessary columns so that the data can be analyzed and visualized correctly.



```python
df = pd.read_csv("titanic.csv") # Load dataset

print(df.head())  # Check first 5 rows

print(df.isnull().sum()) # Check missing values
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df = df.drop(['Name', 'Ticket', 'Cabin'], axis=1)
```

*2. Univariate Analysis 
• Plot the distribution of Age using a histogram and a KDE plot. 
• Plot the count of passengers by Sex using a bar chart. 
• Show the distribution of Fare using a boxplot.*

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)

plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', data=df)

plt.title("Passenger Count by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y='Fare', data=df)

plt.title("Fare Distribution")
plt.ylabel("Fare")

plt.show() 
```

*3. Bivariate Analysis 
• Compare survival rate across Sex using a countplot. 
• Compare Pclass vs Fare using a boxplot. 
• Visualize Age vs Fare using a scatter plot, color-coded by Survived.*

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df)

plt.title("Survival Count by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Pclass', y='Fare', data=df)

plt.title("Passenger Class vs Fare")
plt.xlabel("Pclass")
plt.ylabel("Fare")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)

plt.title("Age vs Fare (Colored by Survival)")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```
